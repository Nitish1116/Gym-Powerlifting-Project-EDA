# Data Visualization

```python
# Amount Lifted by Age
plt.figure(figsize=(12, 10))
sns.histplot(df,x='Age',y='Amount Lifted (kg)', color='blue',kde=True)
plt.title('Amount Lifted by Age')
plt.xlabel('Age (years)')
plt.ylabel('Amount Lifted (kg)')
plt.grid(True)
plt.show()
```

```python
# Boxplot of Amount Lifted by Lift Type
plt.figure(figsize=(12, 6))
sns.boxplot(df,x='Amount Lifted (kg)', y='Lift Type')
plt.title('Amount Lifted by Lift Type')
plt.suptitle('')  # Suppress the default title
plt.xlabel('Amount Lifted (kg)')
plt.ylabel('Lift Type')
plt.show()
```

```python
# Distribution of Lift Types
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='Lift Type', hue='Lift Type')
plt.title('Distribution of Lift Types')
plt.xlabel('Lift Type')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.histplot(df,x='Weight Class',y='Lift Type')
plt.title('Lift Type vs Weight Class')
plt.show()
```

```python
# Average Amount Lifted by Lift Type
avg_lift_by_type = df.groupby('Lift Type').agg({'Amount Lifted (kg)': 'mean'}).reset_index()

plt.figure(figsize=(10, 8))
plt.bar(avg_lift_by_type['Lift Type'], avg_lift_by_type['Amount Lifted (kg)'], color=['blue', 'orange', 'green'])
plt.title('Average Amount Lifted by Lift Type')
plt.xlabel('Lift Type')
plt.ylabel('Average Amount Lifted (kg)')
plt.show()
```

```python
# Average Amount Lifted by Weight Class
avg_lift_by_weight_class = df.groupby('Weight Class').agg({'Amount Lifted (kg)': 'mean'}).reset_index()

plt.figure(figsize=(10, 8))
sns.barplot(x='Weight Class', y='Amount Lifted (kg)', data=avg_lift_by_weight_class, color='k')

plt.title('Average Amount Lifted by Weight Class')
plt.xlabel('Weight Class (kg)')
plt.ylabel('Average Amount Lifted (kg)')
plt.xticks(rotation=45)  # Rotate x-axis labels if necessary for better readability
plt.show()
```

```python
# Performance by Weight Class and Lift Type
plt.figure(figsize=(12, 8))
sns.barplot(data=df, x='Weight Class', y='Amount Lifted (kg)', hue='Lift Type', estimator=sum, errorbar=None)
plt.title('Total Amount Lifted by Weight Class and Lift Type')
plt.ylabel('Total Amount Lifted (kg)')
plt.xlabel('Weight Class (kg)')
plt.legend(title='Lift Type')
plt.show()
```

```python
lift_type_counts = df['Lift Type'].value_counts()
# Plotting the pie chart
plt.figure(figsize=(12, 10))
plt.pie(lift_type_counts, labels=lift_type_counts.index, autopct='%1.1f%%', startangle=90, colors=['red','blue','green'])
plt.title('Distribution of Lift Types')
plt.show()
```

```python
Weight_Class_counts = df['Weight Class'].value_counts()
# Plotting the pie chart
plt.figure(figsize=(12, 10))
plt.pie(Weight_Class_counts, labels=Weight_Class_counts.index, autopct='%1.1f%%', startangle=90, colors=['red','blue','green','yellow','darkgreen','orange','purple','pink','white'])
plt.title('Distribution of Weight Class')
plt.show()

```

```python
g= sns.FacetGrid(df, col="Weight Class", col_wrap=3, height=5)
#col_wrap=2 means that there will be up to 3 plots per row, and height=5 specifies the height of each subplot.
g.map_dataframe(sns.scatterplot, x='Lift Type', y='Amount Lifted (kg)')
g.set_axis_labels('Lift Type', 'Amount Lifted (kg)')
g.add_legend()
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.violinplot(x="Lift Type", y="Amount Lifted (kg)", data=df)

plt.xlabel("Lift Type")
plt.ylabel("Amount Lifted (kg)")
plt.title(" Distribution of lift type")
plt.show()
```
